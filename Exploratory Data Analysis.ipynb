```python
# Fungsi Membuat Kategori Umur
def get_age_range(age):
    if age >= 11 and age <= 26:
        return 'Post Gen Z'
    elif age >= 27 and age <= 42:
        return 'Gen Z'
    elif age >= 43 and age <= 58:
        return 'Millennial'
    elif age >= 59 and age <= 77:
        return 'Pre Boomer'
    else:
        return 'Unknown'

# Membuat kolom baru 'UMUR_RANGE' berdasarkan kolom 'AGE'
data['AGE_RANGE'] = data['UMUR'].apply(get_age_range)
```

```python


# Fungsi Membuat Kategori Pendapatan
def income(AMT_INCOME_TOTAL):
    if (AMT_INCOME_TOTAL < 15935):
        return 'Low Income'
    elif (AMT_INCOME_TOTAL >= 15935 and AMT_INCOME_TOTAL <= 62435):
        return 'Lower Middle Income'
    elif (AMT_INCOME_TOTAL >= 62436 and AMT_INCOME_TOTAL <= 193764):
        return 'Upper Middle Income'
    elif (AMT_INCOME_TOTAL > 193764):
        return 'High Income'

# Mencari Lokasi Kolom INCOME_TYPE
income_loc = income.columns.get_loc(key='AMT_INCOME_TOTAL')

# Menyisipkan Kolom Baru Setelah Kolom AMT_INCOME_TOTAL
data.insert(
    income_loc+1,
    'INCOME_TYPE',
    data['AMT_INCOME_TOTAL'].apply(income)
)
```

```python
#membuat kuartil AMT_GPPDS_PRICE
kuartil_1_income_total= np.percentile(data['AMT_INCOME_TOTAL'], 25)
kuartil_2_income_total= np.percentile(data['AMT_INCOME_TOTAL'], 50)
kuartil_3_income_total= np.percentile(data['AMT_INCOME_TOTAL'], 75)
```

```python
data.loc[(data['AMT_INCOME_TOTAL']<= kuartil_1_income_total), 'range_income_total'] = 'Low income'
data.loc[(data['AMT_INCOME_TOTAL']> kuartil_1_income_total) & (data['AMT_INCOME_TOTAL']<= kuartil_2_income_total ), 'range_income_total'] = 'Lower Middle Income'
data.loc[(data['AMT_INCOME_TOTAL']> kuartil_2_income_total) & (data['AMT_INCOME_TOTAL']<= kuartil_3_income_total ), 'range_income_total'] = 'Upper Middle Income'
data.loc[(data['AMT_INCOME_TOTAL']> kuartil_3_income_total), 'range_income_total'] = 'High Income'
```

```python
# Membuat fungsi untuk mengklasifikasikan resiko kredit
def classify_risk(range_income_total, range_credit):
    if range_income_total == 'low income':
        if range_credit == 'low credit':
            return 'Low Risk'
        elif range_credit == 'mid credit':
            return 'Medium Risk'
        elif range_credit == 'high credit':
            return 'Medium Risk'
        elif range_credit == 'advanced credit':
            return 'High Risk'
    elif range_income_total == 'lower middle income':
        if range_credit == 'low credit':
            return 'Medium Risk'
        elif range_credit == 'mid credit':
            return 'Medium Risk'
        elif range_credit == 'high credit':
            return 'High Risk'
        elif range_credit == 'advanced credit':
            return 'High Risk'
    elif range_income_total == 'upper middle income':
        if range_credit == 'low credit':
            return 'Medium Risk'
        elif range_credit == 'mid credit':
            return 'Medium Risk'
        elif range_credit == 'high credit':
            return 'High Risk'
        elif range_credit == 'advanced credit':
            return 'High Risk'
    elif range_income_total == 'high income':
        if range_credit == 'low credit':
            return 'Medium Risk'
        elif range_credit == 'mid credit':
            return 'High Risk'
        elif range_credit == 'high credit':
            return 'High Risk'
        elif range_credit == 'advanced credit':
            return 'Advanced Risk'
    return 'Unknown'

# Menerapkan fungsi untuk mengklasifikasikan resiko kredit
data['Risk_Category'] = data.apply(lambda x: classify_risk(x['range_income_total'], x['range_credit']), axis=1)

# Menampilkan hasil
print(data[['range_income_total', 'range_credit', 'Risk_Category']])
```
